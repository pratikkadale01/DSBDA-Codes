## Check for null/nan values

```python
data.isnull().sum()
```

### Total null values = 13256
- Car              -  62
- BuildingArea  -   6450
- YearBuilt     -  5375
- CouncilArea   -  1369

```python
# Car - dtpye - float64
# Car - Number of carspots
mode_no=data['Car'].mode()
print('Mode no of car spots = ',mode_no)
print('Total no of NaN values = ',data['Car'].isna().sum())
# Replace NaN values by mean value
data['Car']=data['Car'].fillna(mode_no[0])
print('Total no of NaN values after replacement = ',data['Car'].isna().sum())
```

```python
# BuildingArea - Building Size in Metres
# BuildingArea - dtype - float64
avg_no=data['BuildingArea'].mean()
print('Average value of BuildingArea = ',avg_no)
print('Total no of NaN values = ',data['BuildingArea'].isna().sum())
# Replace NaN values by mean value
data['BuildingArea']=data['BuildingArea'].fillna(avg_no)
print('Total no of NaN values after replacement = ',data['BuildingArea'].isna().sum())
```

```python
# CouncilArea -  Governing council for the area
# CouncilArea - dtype - object
data['CouncilArea'].value_counts()
```

```python
print('Total no of NaN values = ',data['CouncilArea'].isna().sum())
# Drop rows containing nan values
data=data.dropna(subset=['CouncilArea'],axis=0,how='any')
print('Total no of NaN values after dropping rows = ',data['CouncilArea'].isna().sum())
```

```python
# YearBuilt - Year the house was built
# YearBuilt - dtype - float64
print('Total no of NaN values = ',data['YearBuilt'].isna().sum())
# Drop rows containing NaN values
data=data.dropna(subset=['YearBuilt'],axis=0,how='any')
print('Total no of NaN values after replacement = ',data['YearBuilt'].isna().sum())
```

## Data normalization
#### Using Z- Score method (often called standardization) 
- Transforms the info into distribution with a mean of 0 and a typical deviation of 1. 
- Each standardized value is computed by subtracting the mean of the corresponding feature then dividing by the quality deviation.



```python
for col in ['Rooms', 'Price', 'Distance', 'Bedroom2', 'Bathroom', 'Car','Landsize', 'BuildingArea','Propertycount']:
    data[col]=(data[col]-data[col].mean())/data[col].std()
```

# 4. Turn categorical variables into quantitative variables in Python. 

```python
for col in data.select_dtypes(include='object'):
    data[col]=data[col].astype('category')
    data[col]=data[col].cat.codes
```

```python
# One hot encoding strategy is to convert each category value into a new column and assigns a 1 or 0 
# (True/False) value to the column.

data=pd.get_dummies(data, columns=["Type"])
data.head()
```
